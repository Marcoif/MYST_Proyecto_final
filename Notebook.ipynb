> <span style="color:#516D6E; font-family:Georgia; font-size:1.4em;"> Prueba de Heterocedasticidad  </span>


<p style = "font-family: Calibri Light; font-size:1.3em;color:black; LINE-HEIGHT:25px"> Podemos considerar que las observaciones de una variable a lo largo del tiempo (serie de tiempo) presentan heterocedasticidad cuando la varianza que dichas observaciones presentan no es constante a lo largo de dicho periodo temporal.</p>

<p style = "font-family: Calibri Light; font-size:1.3em;color:black; LINE-HEIGHT:25px">Es importante conocer si una serie de tiempo es heterocedástica ya que, por ejemplo, si dicha serie de tiempo presentase esta propiedad, un modelo ARIMA no sería una buena propueta para intentar describir y pronosticar el comportamiento de la varible estudiada.  </p>


<p style = "font-family: Calibri Light; font-size:1.3em;color:black; LINE-HEIGHT:25px"> Para detectar esto, procederemos a utilizar la prueba ARCH (autoregressive conditional heteroskedasticity) con los datos a los que les fue aplicada una diferencia. Para efectos de la prueba, consideraremos, como en el caso de la prueba de estacionariedad, un ${\alpha}=0.05$.</p>



>> <span style="color:#29AECE; font-family:Georgia; font-size:1.2em;"> Prueba Arch </span>
>> <p style = "font-family: Calibri Light; font-size:1.3em;color:black; LINE-HEIGHT:25px"> Supongamos una serie de tiempo del fomra $y_t = \mu_t + \epsilon_t $ donde $\mu_t$ es la media condicional del proceso y $\epsilon_t$ es un término de error con media cero.  </p>

>> <p style = "font-family: Calibri Light; font-size:1.3em;color:black; LINE-HEIGHT:25px"> Si definimos a los resiudales de la serie de tiempo de la siguientes manera : $e_t= y_t-\widehat{mu_t}$
    
   
>> <p style = "font-family: Calibri Light; font-size:1.3em;color:black; LINE-HEIGHT:25px"> La hipótesis alternativa de la prueba está dado por  : $H_a:e_t^2=a_0 + a_ie_{t_i}^2+...+a_me_{t-m}^2+u_t $
    
>> <p style = "font-family: Calibri Light; font-size:1.3em;color:black; LINE-HEIGHT:25px"> Donde $u_t$ es ruido blanco. En cambio la hipótesis nula está dada por : $H_0:a_0 = a_1 =...=a_m=0$ y por tanto, serie de tiempo es homocedástica.
    
    
    
    
    
    




> <span style="color:#516D6E; font-family:Georgia; font-size:1.4em;"> Prueba de normalidad </span>
<p style = "font-family: Calibri Light; font-size:1.3em;color:black; LINE-HEIGHT:25px"> Muchos modelos matemáticos toman como un fuerte supuesto el que los datos a lo cuales se les está aplciando dicho modelo siguen una distribución normal. Si bien en algunos casos esta puede ser una suposición correcta,en muchos otros casos puede no serlo y llevar a que los datos se comporten de manera distinta a lo que sospechamos.</p>


<p style = "font-family: Calibri Light; font-size:1.3em;color:black; LINE-HEIGHT:25px"> En este caso particular se utilizaron dos métodos para conocer si los datos del S&p500 Case-Shiller U.S. National Home Price Index Home Price Index eran normales, los cuales se explican a continuación.</p>

>> <span style="color:#29AECE; font-family:Georgia; font-size:1.2em;"> Diagrama de Cuantil - Cuantil </span>
>><p style = "font-family: Calibri Light; font-size:1.3em;color:black; LINE-HEIGHT:25px"> Este método, nos permite conocer de manera gráfica las diferencias entre una distribución conocida, en este caso la normal, y una distribución empírica. En este gráfica, la linea roja representa la distribución normal y los puntos azules, la distribución empírica de nustros datos.  </p>
>><p style = "font-family: Calibri Light; font-size:1.3em;color:black; LINE-HEIGHT:25px"> Este gráfico contrasta la distribución de los datos en cada uno de sus cuantiles contra una distribución normal, por lo que si nuestros datos fueran normales, estos seguirían una linea 45 grados en el gráfico.  </p>
